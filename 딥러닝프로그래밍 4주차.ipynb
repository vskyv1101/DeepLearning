```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

plt.figure()
plt.scatter(x, y) #산전도
plt.xlabel('Study Time(x)')
plt.ylabel("Score")
plt.grid(True)
plt.show()
```

```python
a = 0
b = 0
#학습률( Learning Rate)
lr = 0.03

#반복시킬 횟수
epoch = 2001

#데이터 개수 세기
n = len(x)
```

```python
#경사 하강법
for i in range(epoch):  #epcoh 만큼 반복(for문 사용)
    y_pred = a * x + b  #예측 시험 점수를 저장
    error = y - y_pred  #실제 값 - 예측 값 (= 오차)
    
    
    a_diff = (2/n) * sum(-x * error)  #오차 함수 MSE를 a로 편미분한 결과
    b_diff = (2/n) * sum(-error) #오차 함수 MSE를 b로 편미분한 결과
    
    #기울기랑 y절편의 값 업데이트 하기
    a = a - (lr * a_diff)
    b = b - (lr * b_diff)
    
    if (i % 100) == 0:
        print("epoch : ", i, ", 기울기 : ", a, ", y-절편: ", b)
```

```python
y_pred = a * x + b
print("예측점수 : ", y_pred)
```

```python
plt.figure()
plt.scatter(x,y, marker='^', s = 200, c="k")
plt.plot(x, y_pred, marker='o', markersize = 10, c ='pink')
plt.xlabel("Study Time(x)")
plt.ylabel("Score(y)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection ='3d')
ax.scatter3D(x1 ,x2, y)
ax.set_xlabel("Study Time (x1)")
ax.set_ylabel("Private Lesson(x2)")
ax.set_zlabel("Study (y)")
plt.show()
```

```python
a1 = 0
a2 = 0
b = 0

lr = 0.01

epoch = 2001

n = len(x1)

for i in range(epoch):
    y_pred = a1 * x1 + a2 * x2 + b #예측
    error = y - y_pred  # 오차
    
    a1_diff = (2/n) * sum(-x1 * error)  # MSE를 a1로 편미분한 결과
    a2_diff = (2/n) * sum(-x2 * error)  # MSE를 a2로 편미분한 결과
    b_diff = (2/n) * sum(-error)         # MSE를 b로 편미분한 결과
    
    #업데이트 하기
    a1 = a1 - lr * a1_diff
    a2 = a2 - lr * a2_diff
    b = b - lr * b_diff
    
    if( i % 100) == 0:
        print("epoch: ", i, ", 기울기1 : ", a1, ", 기울기2 : ", a2, ", 절편 : ", b)

    
    
```

```python
y_pred = a1 * x1 + a2 * x2 + b

print("실제 점수 : ", y)
print("예측 점수 : ", y_pred)
```
